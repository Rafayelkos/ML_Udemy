# Selection of eps (i.e epsilon)

### NearestNeighbors

* Calculate the distance from each point to its closest neighbour using the NearestNeighbor
* The poin itsself included in n_neighbors.
* The kneighbors method returns two arrayes, one which contains the distance to the closest n_neighbors points and the other     which contains the index for each of those points

```python
neigh = NearestNeighbors(n_neighbors =2)
nbrs = neigh.fit(X)
distances, indices = nbrs.kneighbors(X)
```

```python
distances = np.sort(distances, axis = 0)
distances = distances[:,1]
plt.plot(distances)
```

### The optimal value for epsilon will be found at the point of the maximum curvature.

```python
# selecting 0.3 for eps and setting min_samples to 5
m = DBSCAN(eps =0.3, min_samples = 5)
m.fit(X)
```

```python
# The labels_property contains the list of clusters and their respective points.
clusters = m.labels_
```
